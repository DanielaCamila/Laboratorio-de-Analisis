```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('customer_data.csv')
data
```

```python
#Agrupamos los datos por día de la semana y calculamos la media de la factura total por día Creamos un gráfico de barras para visualizar la media de la factura total por día
data.groupby('Segment')['Customer_ID'].count().plot(kind='bar', color='skyblue')

# Añadimos un título al gráfico
plt.title('Cantidad de clientes en cada segmento')

# Etiquetamos el eje X
plt.xlabel('Segment')

# Etiquetamos el eje Y
plt.ylabel('Customer_ID')

# Mostramos el gráfico
plt.show()
```

```python
#Agrupamos los datos por día de la semana y calculamos la media de la factura total por día Creamos un gráfico de barras para visualizar la media de la factura total por día
data.groupby('Segment')['Expected_Lifespan_Months'].mean().plot(kind='bar', color='pink')

# Añadimos un título al gráfico
plt.title('Que tanto se van a quedar según su segmento')

# Etiquetamos el eje X
plt.xlabel('Segment')

# Etiquetamos el eje Y
plt.ylabel('Expected_Lifespan_Months')

# Mostramos el gráfico
plt.show()
```

```python
#Agrupamos los datos por día de la semana y calculamos la media de la factura total por día Creamos un gráfico de barras para visualizar la media de la factura total por día
data.groupby('Segment')['Technical_Issues'].sum().plot(kind='bar', color='purple')

# Añadimos un título al gráfico
plt.title('Según el segmento que tantos erros tecnicos marcaron')

# Etiquetamos el eje X
plt.xlabel('Segment')

# Etiquetamos el eje Y
plt.ylabel('Technical_Issues')

# Mostramos el gráfico
plt.show()
```

### Conclusiones y recomendaciones para mejorar la captación de clientes

- Los Senior son los que peor califican a la empresa
- Los que mejor califican son los Young Professional
- Los que más errores han marcado son los Mid-Career
- Donde hay más clientes es Mid-Career y el número de clientes baja drásticamente con los Senior

Aqui se nota claramente que cuanto más tiempo pasa la gente en la empresa es más probable que terminen calificando peor 

#### Solución que propongo
- Mejorar bastante la experiencia de los Senior para que den una mejor calificación y esten más satisfechos con el servicio
- E igual buscar que los erroes técnicos de los Mid-Career disminuyan, ya que seguramente por eso nos dejan y nunca se pasan a Senior

```python
# Agrupamos los datos por día de la semana y sumamos las facturas totales por día
# Creamos un gráfico de líneas para visualizar la tendencia de la factura total a lo largo de la semana
data.groupby('Acquisition_Month')['Customer_ID'].count().plot(kind='line', marker='o', color='red')

# Añadimos un título al gráfico
plt.title('Cantidad de personas que obtuvieron durante los años')

# Etiquetamos el eje X
plt.xlabel('Acquisition_Month')

# Etiquetamos el eje Y
plt.ylabel('Customer_ID')

# Mostramos el gráfico
plt.show()
```

```python
#Agrupamos los datos por día de la semana y calculamos la media de la factura total por día Creamos un gráfico de barras para visualizar la media de la factura total por día
data.groupby('Segment')['CAC'].mean().plot(kind='bar', color='orange')

# Añadimos un título al gráfico
plt.title('El costo por traer un nuevo cliente ')

# Etiquetamos el eje X
plt.xlabel('Segment')

# Etiquetamos el eje Y
plt.ylabel('CAC')

# Mostramos el gráfico
plt.show()
```

```python
#Agrupamos los datos por día de la semana y calculamos la media de la factura total por día Creamos un gráfico de barras para visualizar la media de la factura total por día
data.groupby('Segment')['Expected_LTV_in_Two_Years'].mean().plot(kind='bar', color='yellow')

# Añadimos un título al gráfico
plt.title('El beneficio que te da cada sección del cliente')

# Etiquetamos el eje X
plt.xlabel('Segment')

# Etiquetamos el eje Y
plt.ylabel('Expected_LTV_in_Two_Years')

# Mostramos el gráfico
plt.show()
```

```python
data['LTV/CAC'] = data['Expected_LTV_in_Two_Years']/data['CAC']
```

```python
data
```

```python
hp = data.groupby(['Segment', 'Acquisition_Month'])['LTV/CAC'].mean().reset_index()
```

```python
# Creamos un gráfico de cajas para visualizar la distribución de la factura total por día de la semana
sns.lineplot(x='Acquisition_Month', y='LTV/CAC', data=hp, hue='Segment')

# Añadimos un título al gráfico
plt.title('Según por segmento que que tanto Litime Value sobre CAC a través del tiempo')

# Mostramos el gráfico
plt.show()
```

### Conclusiones y Recomendaciones
- El mejor grupo a invertir es en el Young Professional	
- En los tres grupos se gasta lo mismo 

#### Conclusiones
- Inviertan en Young Professional
